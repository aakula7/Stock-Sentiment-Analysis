```python
# the CNBC news is scraped using api - 
# https://search.cnbc.com/rs/search/all/view.js?tickersymbols=AAPL&partnerId=10&pagesize=50&pubfreq=a

#The output retrieved is stored in a json file


from nltk.corpus import *
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import nltk
import json
import pandas
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import requests
from nltk.tokenize import *

df = pandas.DataFrame(columns = ['links', 'headline', 'description', 'timeline', 'article'])

stop_words = nltk.corpus.stopwords.words('english')
link = []
headline = []
desc = []
time = []
article_text = []

article_str = '\n'
lst = []

page = ''

with open('cnbc.json') as text:
    response = json.loads(text.read())
    page = response
    
analyser = SentimentIntensityAnalyzer()


for i in page['rss']['channel']['item']:
    link.append(i['link'])
    headline.append(i['title'])
    desc.append(i['description'])
    time.append(i['pubDate'])
    
    inspect_link = urlopen(i['link']).read()
    soup = BeautifulSoup(inspect_link, 'lxml')
    text = soup.find('div', class_="ArticleBody-articleBody")
    if text is not None:
        article_str = ' '
        article = text.find_all('p')
        for t in article:
            article_str += str(t)
        cleanr = re.compile('<.*?>')
        clean_article = re.sub(cleanr, ' ', article_str)
        article_text.append(clean_article)
       
    else:
        article_text.append('None')
        
article_new = []        
for i in article_text:
    i = i.lower()
    i = i.lstrip(' ')
    sentence = ' '
    word_tokens = word_tokenize(i)
    for w in word_tokens:
        if w not in stop_words:
            w = re.sub('[^.,a-zA-Z0-9 \n\.]', ' ', w)
            
            sentence += w
            sentence += ' '
    article_new.append(sentence)
    sentence = article_new
    score = analyser.polarity_scores(str(sentence))
    print(article_new)
    print(score)
    break
            
        
'''df['links'] = link
df['headline'] = headline
df['description'] = desc
df['timeline'] = time
df['article'] = article_text  
df.to_csv(r'CNBC_articles.csv')'''

```
