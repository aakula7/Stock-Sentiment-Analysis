```python
driver = webdriver.Edge(executable_path = 'C:/Users/aakul/Downloads/msedgedriver.exe')
driver.get('https://www.wsj.com/market-data/quotes/AAPL')
other_page = driver.page_source
other_soup = BeautifulSoup(other_page, 'lxml')
posts = other_soup.findAll('li')

more_news = list(posts)

x = 1
while(x < 40):
    #driver.execute_script('windows.scrollTo(0, document.body.scrollHeight);')
    try:
        time.sleep(1.5)
        btn = driver.find_element_by_xpath("/html/body/div[2]/div[3]/section[2]/div[2]/div[1]/div/div/div[2]/a")
        
        print('btn[load more]:', btn, '\n')
        
        ActionChains(driver).move_to_element(btn).click(btn).perform()
        print('btn is clicked')
        
        clicked_page = driver.page_source
        clicked_soup = BeautifulSoup(clicked_page, 'lxml')
        clicked_posts = clicked_soup.findAll('li')
        
        more_news.append(clicked_posts)
        
        x += 1
        
    except Exception as e:
        print('Click Error:', e)
        
    #click = input('Want to click more? (Y,N)')
    #print(click)
    #if click != 'Y':
        #break
        
driver.quit()
```
